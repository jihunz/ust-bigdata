# 1. One-hot encoding 연습: edge 데이터중 범주형 데이터(weight_sign 칼럼/feature)를 One-hot encoding하여 수치형 데이터로 변환하시오.

```python
import pandas as pd
```

```python
edges = pd.DataFrame({'source': [0, 1, 2], 'target': [2, 2, 3],
                      'weight': [3, 4, 5], 'color': ['red', 'blue', 'blue']})
edges
```

```python
weight_dict = {3:"M", 4:"L", 5:"XL"}
edges["weight_sign"] = edges["weight"].map(weight_dict)
```

```python
edges
```

```python
weight_sign = pd.get_dummies(edges) # 이 부분을 완성하여, 다음과 같은 결과가 출력되도록 하시오.
weight_sign
```
