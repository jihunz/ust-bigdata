* Weight 와 데이터 X 의 곱셈 구하기: 간단한 예제

```python
import numpy as np # Numpy 라이브러리 호출
```

```python
weight_vector = np.array([[1], [1], [1]]) # Weight Vector
print(weight_vector)
print(weight_vector.shape)
```

```python
print(weight_vector.T)
print(weight_vector.T.shape)
```

```python
x_vector = np.array([[3], [4], [5]])
print(x_vector)
print(x_vector.shape)
```

* (1,3)*(3,1) 의 곱셈은 (1,1)

```python
#numpy에서 dot 함수는 2차원 배열(행렬)일 행렬 곱셈을 수행합니다. 1차원 배열(벡터): 벡터의 내적 계산
weight_vector.T.dot(x_vector) # 1 * 3 + 1 * 4 + 1 * 5 = 12
```

# 실제 보스턴 집값 데이터와 Weight의 곱셈의 예
 - 모든 행의 각 데이터와 weight값의 곱셈

image.png

```python
import pandas as pd
```

```python
#data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data' #Data URL
data_url = './data/housing.data' #Data URL
df_data = pd.read_csv(data_url, sep='\\s+', header = None) #csv 타입 데이터 로드, separate는 빈공간으로 지정하고, Column은 없음
df_data.columns = ['CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV']
df_data.head()
```

```python
df_data['weight_0'] = 1 # weight 0 값 추가
df_data= df_data.drop("MEDV", axis=1) #Y값 제거
df_data.head()
```

```python
df_matrix = df_data.to_numpy() # Matrix Data로 변환하기
print(df_matrix)
```

```python
print(df_matrix.shape)
```

```python
# weight를 random하게 생성
weight_vector =  np.random.random_sample((14, 1))
print(weight_vector)
```

```python
print(weight_vector.shape)
```

* 데이터 (506, 14) \* weight (14, 1) = (506 * 1)

```python
# shape (506, 14) * (14, 1) = (506 * 1)
df_matrix.dot(weight_vector)
```

```python
print(df_matrix.dot(weight_vector).shape)
```
